```python
import cv2
from matplotlib import pyplot as plt
import numpy as np
```

```python
detector = cv2.CascadeClassifier('haarcascade_russian_plate_number.xml')
```

```python
def plt_show(image, title="", gray = False, size =(100,100)):
    temp = image
    if gray == False:
        temp = cv2.cvtColor(temp, cv2.COLOR_BGR2RGB)
        plt.title(title)
        plt.imshow(temp, cmap='gray')
        plt.show()
        
```

```python
def detect_number(img):
    temp = img
    gray = cv2.cvtColor(temp, cv2.COLOR_BGR2GRAY)
    number = detector.detectMultiScale(img,1.2)
    print("number of number plates detect:"+str(len(number)))
    for numbers in number:
        (x,y,w,h) = numbers
        roi_gray = gray[y:y+h, x:x+w]
        roi_color = img[y:y+h, x:x+h]
        cv2.rectangle(temp, (x,y), (x+w,y+h), (0,255,0), 3)
        
    plt_show(temp)
```

```python
img = cv2.imread("car_4.jpg")
plt_show(img)
detect_number(img)
```
